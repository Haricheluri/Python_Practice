Exercise 4: Calculator Discount A store offers:

10% discount if purchase > $100.

5% discount if purchase > $50.

No discount otherwise.

Calculate the final price.

```python
price = float(input("Enter purchase amount: "))

if price > 100:
    discount = price * 0.10
elif price > 50:
    discount = price * 0.05
else:
    discount = 0

final_price = price - discount
print("Final price:", final_price)
```

Exercise 6: Grade Calculator with Bonus Calculate grades (A-F) but add +1 bonus grade level if attendance >
90%.

```python
marks = int(input("Enter marks: "))
attendance = int(input("Enter attendance: "))

if marks >= 90:
    grade = 'A'
elif marks >= 80:
    grade = 'B'
elif marks >= 70:
    grade = 'C'
elif marks >= 60:
    grade = 'D'
else:
    grade = 'F'

# Bonus for attendance
if attendance > 90 and grade != 'A' and grade != 'F':
    grade = chr(ord(grade) - 1)

print("Final Grade:", grade)
```

Exercise 7: Login System with Attempts Create a login system that:

Checks username and password

Tracks login attempts (max 3)

Locks account after failed attempts

```python
correct_username = "admin"
correct_password = "1234"

attempts = 0
max_attempts = 3

while attempts<max_attempts:
    username=input("Enter Username: ")
    password=input("Enter Password: ")
    if username==correct_username and password==correct_password:
        print("Login Successful")
        break
    else:
        attempts+=1
        print("Wrong username")
        print(f"No of attempts left {max_attempts-attempts}")
else:
    print("Too many attempts")
```

Exercise 8: Shipping Cost Calculator Calculate shipping based on:

Weight of package

Destination (domestic/international)

Express or standard shipping

```python
weight = float(input("Enter package weight (kg): "))
destination = input("Enter destination (domestic/international): ").lower()
shipping_type = input("Enter shipping type (standard/express): ").lower()
domestic=100
international=200
cost=0
if destination=="domestic":
    if shipping_type=="express":
        cost=weight*domestic*50
    elif shipping_type=='standard':
        cost=weight*domestic*10
    else:
        print("Invalid shipping Type")
elif destination=="international":
    if shipping_type=="express":
        cost=weight*international*50
    elif shipping_type=='standard':
        cost=weight*international*10
    else:
        print("Invalid shipping Type")
else:
    print("Invalid destination")

if cost>0:
    print(f"shipping cost  for the destination: {destination} and shipping type: {shipping_type} is {cost}")
```
