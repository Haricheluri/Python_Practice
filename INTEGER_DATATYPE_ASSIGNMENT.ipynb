Calculate the sum of digits in number 12345

```python
num=12345
sum1=0
while num!=0:
    sum1+=num%10
    num=num//10
print(f"sum of digits in a number 12345 :{sum1}")
```

Check if 97 is a prime number

```python
num=17
count=0
for i in range(2,int(num**0.5)):
    if num%i==0:
        count+=1
if count==0:
    print("Prime")
else:
    print("Not prime")
```

Question 7: Find the factorial of 8

```python
num=8
fact=1
for i in range(2,num+1):
    fact*=i
print(f"factorial of a number : {fact}")
```

Question 8: Calculate the average of numbers: 15, 23, 31, 42, 56

```python
numbers=[15,23,31,42,56]
average=sum(numbers)//len(numbers)
print(f"Average of the numbers : {average}")
```

Find the greatest common divisor (GCD) of 48 and 36

```python
import math 
num1=36
num2=48
print(f"Gcd of the numbers: {math.gcd(num1,num2)}")
```

```python
num1=36
num2=48
while num2!=0:
    num1,num2=num2,num1%num2
print(f"Gcd of the numbers :{num1}")
```
