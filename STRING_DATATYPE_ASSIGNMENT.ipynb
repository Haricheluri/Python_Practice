```python
s1 = "programming"
s2 = "grammar"

longest = ""

for i in range(len(s1)):
    for j in range(i + 1, len(s1) + 1):
        sub = s1[i:j]
        if sub in s2 and len(sub) > len(longest):
            longest = sub

print("Longest Common Substring:", longest)

```

Question 46: Convert "hello world" to Pig Latin

If a word starts with a consonant, move the first letter to the end and add “ay”

If a word starts with a vowel, add “way” at the end

```python
text = "hello world"
words = text.split()

vowels = "aeiouAEIOU"
result = []

for word in words:
    if word[0] in vowels:
        pig_word = word + "way"
    else:
        pig_word = word[1:] + word[0] + "ay"
    result.append(pig_word)

print(" ".join(result))
```

Question 47: Check if string is a valid IPv4 address: "192.168.1.1"

```python
ip = "192.168.1.1"

parts = ip.split(".")

valid = True

if len(parts) != 4:
    valid = False
else:
    for part in parts:
        if not part.isdigit():  #check digits
            valid = False
            break

        if len(part) > 1 and part[0] == "0":  #leading zeros
            valid = False
            break

        num = int(part)
        if num < 0 or num > 255:
            valid = False
            break

if valid:
    print("Valid IPv4 address")
else:
    print("Invalid IPv4 address")

```

Question 49: Convert "hello world" to ROT13 encoding


ROT13 = shift each letter by 13 places in the alphabet

```python
text = "hello world"
result = ""

for ch in text:
    if 'A' <= ch <= 'Z':
        result += chr((ord(ch) - ord('A') + 13) % 26 + ord('A'))
    elif 'a' <= ch <= 'z':
        result += chr((ord(ch) - ord('a') + 13) % 26 + ord('a'))
    else:
        result += ch

print(result)
```

Question 50: Check if string is a valid credit card number: "4532015112830366"



Count total digits = n

If n is even
→ double digits at even positions (0, 2, 4 …)

If n is odd
→ double digits at odd positions (1, 3, 5 …)

If doubled value > 9 → subtract 9

Add all digits

If total % 10 == 0 → Valid

```python
card = "4532015112830366"
total = 0
n = len(card)

for i in range(n):
    d = int(card[i])

    if (n % 2 == 0 and i % 2 == 0) or (n % 2 == 1 and i % 2 == 1):
        d = d * 2
        if d > 9:
            d = d - 9

    total = total + d

if total % 10 == 0:
    print("Valid")
else:
    print("Invalid")
```
